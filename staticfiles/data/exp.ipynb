```python
import pandas as pd
import hashlib
import string
import random

# ------------------------------------------------
# Generate UNIQUE PAN Card per row
# ------------------------------------------------
def generate_pan(student_name, row_number):
    # Random 3-letter prefix (A–Z)
    prefix = ''.join(random.choices(string.ascii_uppercase, k=3))

    category = "P"  # Individual

    # 5th character: First letter of surname
    surname_letter = str(student_name).strip().split()[-1][0].upper()

    # Numeric part (0001–9999)
    numeric_part = f"{row_number:04d}"

    # Base PAN
    base_pan = f"{prefix}{category}{surname_letter}{numeric_part}"

    # Hash-based check digit (A–Z)
    hash_val = int(hashlib.md5(base_pan.encode()).hexdigest(), 16)
    check_digit = string.ascii_uppercase[hash_val % 26]

    return base_pan + check_digit


# ------------------------------------------------
# Process CSV File
# ------------------------------------------------
def process_university_csv(input_file, output_file):
    df = pd.read_csv(input_file)

    pan_set = set()
    pan_cards = []

    for i, name in enumerate(df["Name"], start=1):
        # Loop until a truly unique PAN is generated
        while True:
            pan = generate_pan(name, i)
            if pan not in pan_set:
                pan_set.add(pan)
                pan_cards.append(pan)
                break

    df["PAN_CARD"] = pan_cards
    df.to_csv(output_file, index=False)

    print(f"✔ Unique PAN_CARD generated for {input_file}")


# ------------------------------------------------
# Run for all universities
# ------------------------------------------------
process_university_csv("KL_University_Students.csv", "KL_University_Students_PAN.csv")
process_university_csv("LPU_University_Students.csv", "LPU_University_Students_PAN.csv")
process_university_csv("MBU_University_Students.csv", "MBU_University_Students_PAN.csv")
process_university_csv("SRM_University_Students.csv", "SRM_University_Students_PAN.csv")
```

```python
import pandas as pd

# ------------------------------------------------
# Add password column to existing CSV
# ------------------------------------------------
def add_password_column(input_file, output_file):
    df = pd.read_csv(input_file)

    # Create password: StudentID + last 4 chars of PAN_CARD
    df["password"] = (
        df["StudentID"].astype(str) +
        df["PAN_CARD"].astype(str).str[-4:]
    )

    df.to_csv(output_file, index=False)
    print(f"✔ Password column added to {output_file}")


# ------------------------------------------------
# Apply to all generated CSVs
# ------------------------------------------------
add_password_column(
    "KL_University_Students_PAN.csv",
    "KL_University_Students_PAN_PWD.csv"
)

add_password_column(
    "LPU_University_Students_PAN.csv",
    "LPU_University_Students_PAN_PWD.csv"
)

add_password_column(
    "MBU_University_Students_PAN.csv",
    "MBU_University_Students_PAN_PWD.csv"
)

add_password_column(
    "SRM_University_Students_PAN.csv",
    "SRM_University_Students_PAN_PWD.csv"
)
```
